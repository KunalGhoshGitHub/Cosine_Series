# Individual Terms of Cosine Series

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython import display
from math import factorial,sin

%matplotlib notebook

n = 4
for p in range(0,n,1):
    k = (2*p)
    fig, ax = plt.subplots(figsize=(10,5.625))

    # Set the range for the axis
    ax.set_xlim(-2*np.pi, 2*np.pi)
    ax.set_ylim(-2, 2)
    # The x axis and y axis
    # x axis
    plt.plot([-4*np.pi,4*np.pi],[0,0],color = "#d9d9d9")
    # y axis
    plt.plot([0,0],[-2,2],color = "#d9d9d9")

    # The function you want to draw
    def f(i):
        y = ((-1)**p)*(i**k)/factorial(k)
        return y
    
    def f1(i):
        y = np.cos(i)
        return y

    def animation_frame(i):
        x_data = np.linspace(-i*np.pi/180,i*np.pi/180,i)
        y_data = f(x_data)
        y = f1(x_data)
        # ax.set_facecolor('xkcd:mint green')
        # This is the graph to be displayed
        plt.plot(x_data,y_data,"g",label = "x")
        plt.plot(x_data,y,"m",label = "sin(x)")
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        name = "(x^"+str(k)+")/"+"("+str(k)+"!)"
        if (p%2) != 0:
            name = "-"+name
        plt.legend(["","",name,"cos(x)"])

    
    anim = FuncAnimation(fig, func=animation_frame, frames=range(0, 360+1, 1), interval=5)
    plt.show()
    writervideo = animation.FFMpegWriter(fps=50) #fps = Frame per second (This is the speed of the  animation) 
    file_name = "Term_"+str(p+1)+"_of_cos.mp4"
    anim.save(file_name, writer=writervideo, dpi = 250) #dpi = dots per inch (This is the quality of the image)
```

# Sum of the first n terms of the Cosine Series:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython import display
from math import factorial,sin

%matplotlib notebook

n = 7
for p in range(0,n,1):
    #print(p)
    k = (2*p)
    fig, ax = plt.subplots(figsize=(10,5.625))
    
    angle = 360

    # Set the range for the axis
    ax.set_xlim(-(angle/180)*np.pi, (angle/180)*np.pi)
    ax.set_ylim(-2, 2)
    # The x axis and y axis
    # x axis
    plt.plot([-(angle/180)*np.pi,(angle/180)*np.pi],[0,0],color = "#d9d9d9")
    # y axis
    plt.plot([0,0],[-2,2],color = "#d9d9d9")

    # The function you want to draw
    def f(i):
        y = 0
        for j in range(0,p+1,1):
            h = (2*j)
            y = y + ((-1)**j)*(i**(h))/factorial(h)
        return y
    
    def f1(i):
        y = np.cos(i)
        return y

    def animation_frame(i):
        x_data = np.linspace(-i*np.pi/180,i*np.pi/180,i)
        y_data = f(x_data)
        y = f1(x_data)
        # ax.set_facecolor('xkcd:mint green')
        # This is the graph to be displayed
        plt.plot(x_data,y_data,"g",label = "x")
        plt.plot(x_data,y,"m",label = "cos(x)")
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        Name = "Sum of "+str(p+1) +" Term(s)"
        plt.legend(["","",Name,"cos(x)"])


    anim = FuncAnimation(fig, func=animation_frame, frames=range(0, angle+1, 1), interval=5)
    plt.show()
    writervideo = animation.FFMpegWriter(fps=50) #fps = Frame per second (This is the speed of the  animation) 
    file_name = "Sum_of_"+str(p+1)+"_terms_of_cos"+".mp4"
    anim.save(file_name, writer=writervideo, dpi = 250) #dpi = dots per inch (This is the quality of the image)
```
